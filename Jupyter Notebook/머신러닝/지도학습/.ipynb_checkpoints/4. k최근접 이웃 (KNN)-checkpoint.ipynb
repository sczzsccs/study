# 최근접 이웃(K-Nearest Neighbor)

* 특별한 예측 모델 없이 가장 가까운 데이터 포인트를 기반으로 예측을 수행하는 방법
* 분류와 회귀 모두 지원

![k nearest neighbor](https://upload.wikimedia.org/wikipedia/commons/thumb/e/e7/KnnClassification.svg/220px-KnnClassification.svg.png)

* k개의 데이터 포인트 중 가장 많은 클래스가 예측 결과

```python
# 데이터 시각화 함수
import numpy as np
def make_meshgrid(x, y, h=0.02):
    x_min, x_max = x.min()-1, x.max()+1
    y_min, y_max = y.min()-1, y.max()+1
    xx, yy =np.meshgrid(np.arange(x_min, x_max, h),
                        np.arange(y_min, y_max, h))
    return xx, yy

def plot_contours(clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = plt.contourf(xx, yy, Z, **params)
    return out
```

```python
tsne = TSNE(n_components=2)
X_comp = tsne.fit_transform(X) # 저차원 변환

iris_comp = pd.DataFrame(X_comp)
iris_comp['target'] = y
iris_comp
```

```python
iris_comp.describe()
```

```python
plt.scatter(X_comp[:, 0], X_comp[:, 1], c=y, cmap=plt.cm.autumn, s=20, edgecolors='g')
```

```python
model = KNeighborsClassifier()
model.fit(X_comp, y)
model.predict(X_comp)

xx, yy = make_meshgrid(X_comp[:, 0], X_comp[:, 1])
plot_contours(model, xx, yy, cmap=plt.cm.autumn, alpha=0.8)
plt.scatter(X_comp[:, 0], X_comp[:, 1], c=y, cmap=plt.cm.Accent, s=20, edgecolors='r')
```

```python
# 스케일 전
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)

print(f"스케일 안한 학습 데이터: {model.score(X_train, y_train)}")
print(f"스케일 안한 평가 데이터: {model.score(X_test, y_test)}")

# 스케일 후
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.fit_transform(X_test)

model = KNeighborsClassifier()
model.fit(X_train_scl, y_train)

print(f"스케일 한 학습 데이터: {model.score(X_train, y_train)}")
print(f"스케일 한 평가 데이터: {model.score(X_test, y_test)}")
print(f"스케일 학습 데이터: {model.score(X_train_scl, y_train)}")
print(f"스케일 평가 데이터: {model.score(X_test_scl, y_test)}")
```

```python
from sklearn.pipeline import make_pipeline
esti = make_pipeline(
    StandardScaler(), KNeighborsClassifier()
)

from sklearn.model_selection import cross_validate
import multiprocessing
cross_validate(
    estimator=esti,
    X=X, y=y, cv=5,
    n_jobs=multiprocessing.cpu_count(),
    verbose=True)
```

```python
from sklearn.pipeline import Pipeline
Pipe = Pipeline(
    [('scaler', StandardScaler()),
     ('model', KNeighborsClassifier())])

parm = [{'model__n_neighbors': [3, 5, 7],
         'model__weights': ['uniform', 'distance'],
         'model__algorithm': ['ball_tree', 'kd_tree', 'brute']}]

from sklearn.model_selection import GridSearchCV
import multiprocessing
gs = GridSearchCV(
    estimator=Pipe,
    param_grid=parm,
    n_jobs=multiprocessing.cpu_count(),
    verbose=True)
gs.fit(X, y)
```

```python
gs.best_estimator_
```

```python
print(f'GridSearchCV best score: {gs.best_score_}')
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
X_comp = tsne.fit_transform(X) # 저 차원 변환

cancer_comp = pd.DataFrame(X_comp)
cancer_comp['target'] = y
cancer_comp
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
scaler = StandardScaler()
X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.fit_transform(X_test)
model = KNeighborsRegressor()
model.fit(X_train_scl, y_train)
print(f"스케일 학습 데이터: {model.score(X_train_scl, y_train)}")
print(f"스케일 평가 데이터: {model.score(X_test_scl, y_test)}")

X, y = X_train_scl, y_train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler
esti = make_pipeline(MinMaxScaler(), KNeighborsRegressor())

from sklearn.model_selection import cross_validate
import multiprocessing
cross_validate(estimator=esti, X=X, y=y, cv=5,
               n_jobs=multiprocessing.cpu_count(), verbose=True)
```

```python
from sklearn.pipeline import Pipeline
Pipe = Pipeline([('scaler', MinMaxScaler()),
                 ('model', KNeighborsRegressor())])
parm = [{'model__n_neighbors': [3, 5, 7],
         'model__weights': ['uniform', 'distance'],
         'model__algorithm': ['ball_tree', 'kd_tree', 'brute']}]

from sklearn.model_selection import GridSearchCV
import multiprocessing
gs = GridSearchCV(estimator=Pipe,
                  param_grid=parm,
                  n_jobs=multiprocessing.cpu_count(),
                  verbose=True)
gs.fit(X, y)
print(f'GridSearchCV best params: {gs.best_params_}')
print(f'GridSearchCV best estimator: {gs.best_estimator_}')
print(f'GridSearchCV best score: {gs.best_score_}')
```
