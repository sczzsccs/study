* 회귀, 분류, 이상치 탐지 등에 사용되는 지도학습 방법
* 클래스 사이의 경계에 위치한 데이터 포인트를 서포트 벡터(support vector)라고 함
* 각 지지 벡터가 클래스 사이의 결정 경계를 구분하는데 얼마나 중요한지를 학습
* 각 지지 벡터 사이의 마진이 가장 큰 방향으로 학습
* 지지 벡터 까지의 거리와 지지 벡터의 중요도를 기반으로 예측을 수행

![support vector machine](https://upload.wikimedia.org/wikipedia/commons/thumb/2/20/Svm_separating_hyperplanes.png/220px-Svm_separating_hyperplanes.png)

* H3은 두 클래스의 점들을 제대로 분류하고 있지 않음
* H1과 H2는 두 클래스의 점들을 분류하는데, H2가 H1보다 더 큰 마진을 갖고 분류하는 것을 확인할 수 있음

### SVM을 사용한 회귀 모델 (SVR)

입력 데이터를 고차원 공간에 사상해서 비선형 특징을 학습할 수 있도록 확장하는 방법
scikit-learn에서는 Linear, Polynomial, RBF(Radial Basis Function)등 다양한 커널 기법을 지원

```python
from sklearn.datasets import make_regression ## 보스턴 대타 회귀데이터 생성용 모듈
# 회귀분석용 1000개의 데이터(sample), 1개의 특성(feature)을 가진 가상 데이터 셋 생성.
X, y = make_regression(n_samples = 1000, n_features = 1, bias = 1000, noise = 50, random_state=333)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### 데이터 전처리
- SVM은 입력 데이터가 정규화 되어야 좋은 성능을 보임
- 주로 모든 특성 값을 [0, 1] 범위로 맞추는 방법을 사용
- scikit-learn의 StandardScaler 또는 MinMaxScaler를 사용해 정규화

```python
# 전처리 전
from sklearn.datasets import load_breast_cancer
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=123)

from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)

print(f"SVC 전처리 전 학습 데이터 점수: {model.score(X_train, y_train)}")
print(f"SVC 전처리 전 학습 데이터 점수: {model.score(X_test, y_test)}")
```

#### StandardScaler 전처리

```python
# StandardScaler 전처리
from sklearn.preprocessing import StandardScaler
sca = StandardScaler()
X_train = sca.fit_transform(X_train)
X_test = sca.transform(X_test)

from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)

print(f"SVC StandardScaler 학습 데이터 점수: {model.score(X_train, y_train)}")
print(f"SVC StandardScaler 학습 데이터 점수: {model.score(X_test, y_test)}")
```

#### MinMaxScaler 전처리

```python
# MinMaxScaler 전처리
from sklearn.preprocessing import MinMaxScaler
sca = MinMaxScaler()
X_train = sca.fit_transform(X_train)
X_test = sca.transform(X_test)

from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)

print(f"SVC MinMaxScaler 학습 데이터 점수: {model.score(X_train, y_train)}")
print(f"SVC MinMaxScaler 학습 데이터 점수: {model.score(X_test, y_test)}")
```

### 그래프 표현을 위한 함수

```python
def make_mashgrid(x, y, h=0.2):
    x_min, x_max = x.min()-1, x.max()+1
    y_min, y_max = y.min()+1, y.max()+1
    x_range, y_range = np.meshgrid(np.arange(x_min, x_max, h),
                                   np.arange(y_min, y_max, h))
    return x_range, y_range

def plot_contours(cl_m, x_range, y_range, **prams):
    z = cl_m.predict(np.c_[x_range.ravel(), y_range.ravel()])
    z = z.reshape(x_range.shape)
    out = plt.contourf(x_range, y_range, z, **prams)
    return out
```

```python
from sklearn.manifold import TSNE
X_comp = TSNE(n_components=2).fit_transform(X) # 저 차원변환

X_0, X_1 = X_comp[:, 0], X_comp[:, 1]
x_mash, y_mash = make_mashgrid(X_0, X_1)

import matplotlib.pyplot as plt
model.fit(X_comp, y)
plot_contours(model, x_mash, y_mash, cmap=plt.cm.cool, alpha=0.7)
plt.scatter(X_0, X_1, c=y, cmap=plt.cm.coolwarm, s=20, edgecolors='y')
```
