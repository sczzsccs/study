## model_selection 모듈

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes

diat = load_diabetes()
X_train, X_test, y_train, y_test = train_test_split(
    diat.data, diat.target, test_size=0.3)

model = LinearRegression()
model.fit(X_train, y_train)

print("학습 데이터 점수: ", model.score(X_train, y_train))
print("평가 데이터 점수: ", model.score(X_test, y_test))
```

```python
predict = model.predict(X_test)
expected = y_test
plt.figure(figsize=(8,4))
plt.scatter(predict, expected)
plt.plot([30, 300], [30, 350], "g")
plt.grid()
plt.tight_layout()
```

## cross_val_score(): 교차 검증

```python
from sklearn.model_selection import cross_val_score, cross_validate

scor = cross_val_score(model, diat.data, diat.target, cv=5)

print("교차 검증 정확도: ", scor)
print(f"차 검증 정확도(평균): {np.mean(scor)} +/- {np.std(scor)}")
```

## GridSearchCV: 교차 검증과 최적 하이퍼 파라미터 찾기

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
import pandas as pd

al=[0.001*(10**i) for i in range(7)]
pram_grid = dict(alpha=al)

gs =GridSearchCV(estimator=Ridge(), param_grid=pram_grid, cv=10)
res=gs.fit(diat.data, diat.target)
print("최적점수: ", res.best_score_)
print("최적 파라미터: ", res.best_params_)
print(gs.best_estimator_)
pd.DataFrame(res.cv_results_)
```

## preprocessing 데이터 전처리 모듈

```python
import multiprocessing
from sklearn.datasets import load_iris # 붓꽃 데이터
from sklearn.linear_model import LogisticRegression

iris = load_iris()

parm_grid = [{'penalty': ['none', 'l2'], 
              'C': [1.5, 2.0, 2.5, 3.0, 3.5]}]

gs = GridSearchCV(estimator=LogisticRegression(), 
                  param_grid=parm_grid,
                  scoring='accuracy',
                  cv=10,
                  n_jobs=multiprocessing.cpu_count(),
                  error_score='raise')

res = gs.fit(iris.data, iris.target)

print("최적점수: ", res.best_score_)
print("최적 파라미터: ", res.best_params_)
print(gs.best_estimator_)
pd.DataFrame(res.cv_results_)
```

## 오차 행렬(Confusion Matrix)

* True Negative: 예측값을 Negative 값 0으로 예측했고, 실제 값도 Negative 값 0
* False Positive: 예측값을 Positive 값 1로 예측했는데, 실제 값은 Negative 값 0
* False Negative: 예측값을 Negative 값 0으로 예측했는데, 실제 값은 Positive 값 1
* True Positive: 예측값을 Positive 값 1로 예측했고, 실제 값도 Positive 값 1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

scal = MinMaxScaler() # 정규화
iris_scal = scal.fit_transform(iris_df)
iris_df_scal = pd.DataFrame(data=iris_scal, columns=iris.feature_names)
print(iris_df_scal.describe())
print()

X, y =make_classification(n_samples=1000,
                          n_features=2,
                          n_informative=2,
                          n_redundant=0,
                          n_clusters_per_class=1)

X_train, X_test, y_train, y_test = train_test_split(
    iris_df_scal, iris.target, test_size=0.3)

model = LogisticRegression() # 로지스틱회귀
model.fit(X_train, y_train)

print("훈련 데이터 점수:", model.score(X_train, y_train))
print("평가 데이터 점수:", model.score(X_test, y_test))
print()

################################################################

from sklearn.metrics import confusion_matrix

# 모델예측
predi = model.predict(X_test)
print("predict:\n", predi)
con = confusion_matrix(y_true=y_test, y_pred=predi)
print("\nconfusion_matrix:\n", con)

fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(con, cmap=plt.cm.Blues, alpha=0.3)

for i in range(con.shape[0]):
    for j in range(con.shape[1]):
        ax.text(x=j, y=i, s=con[i, j], va='center', ha='center')

plt.xlabel('predict')
plt.ylabel('True')
plt.tight_layout()
plt.show()
```

### 오차 행렬(Confusion Matrix)

## 정밀도(Precision)와 재현율(Recall)

- 정밀도 = TP / (FP + TP) 예측이 P(1)인 것중 실제가 P(1)인 비
- 재현율 = TP / (FN + TP) 실제가 P(1)인 것중 예측이 P(1)인 비


* 정확도 = (TN + TP)      / (TN + FP + FN + TP) 
* (파란색(예측이 맞은 부분)/ 전체범위)
* 오류율 = (FN + FP)      / (TN + FP + FN + TP) 
* ( 흰색 (예측이 틀린 부분)/ 전체범위)

- 오류율 + 정확도 = 1(100%)

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris

X, y =make_classification(n_samples=1000,
                          n_features=2,
                          n_informative=2,
                         n_redundant=0,
                         n_clusters_per_class=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_df_scal, iris.target, test_size=0.3)

 # 로지스틱회귀
model = LogisticRegression()
model.fit(X_train, y_train)

print("훈련 데이터 점수:", model.score(X_train, y_train))
print("평가 데이터 점수:", model.score(X_test, y_test))
print()

######## 오차행렬 ########################################################
from sklearn.metrics import confusion_matrix

# 모델예측
predi = model.predict(X_test)
con = confusion_matrix(y_true=y_test, y_pred=predi)
print("\nconfusion_matrix:\n", con)

fig, ax = plt.subplots(figsize=(5, 5))
ax.matshow(con, cmap=plt.cm.Blues, alpha=0.3)

for i in range(con.shape[0]):
    for j in range(con.shape[1]):
        ax.text(x=j, y=i, s=con[i, j], va='center', ha='center')

plt.xlabel('predict')
plt.ylabel('True')
plt.tight_layout()
plt.show()

####### 정밀도, 재현율 ##########################################
from sklearn.metrics import precision_score, recall_score

preci = precision_score(y_test, predi, average="macro")
recall = recall_score(y_test, predi, average="macro")

print("정밀도:", preci)
print("재현율:", recall)
```

## F1 Score

- 정밀도와 재현율을 결합한 지표
- 정밀동와 재현율이 어느 한쪽으로 치우치지 않을 때 높은 값

\begin{equation}
F1 = 2 \times \frac{precision \times recall}{precision + recall}
\end{equation}

```python
from sklearn.metrics import f1_score

f1= f1_score(y_test, predi, average="macro")
f1
```

## ROC 곡선과 AUC

* ROC 곡선은 FPR(False Positive Rate)이 변할 때 TPR(True Positive Rate)이 어떻게 변하는지 나타내는 곡선
  + TPR(True Positive Rate): TP / (FN + TP), 재현율
  + TNR(True Negative Rate): TN / (FP + TN)
  + FPR(False Positive Rate): FP / (FP + TN), 1 - TNR

* AUC(Area Under Curve) 값은 ROC 곡선 밑에 면적을 구한 값 (1이 가까울수록 좋은 값)

```python
from sklearn.metrics import roc_curve

score1  = model.predict_proba(X_test)[:, 1]
fprs, tprs, thres = roc_curve(y_test, score1, pos_label=1)

plt.plot(fprs, tprs, label='ROC')
plt.plot([0, 1], [0, 1], label="Random")
plt.legend()
start, end = plt.xlim()
plt.xticks(np.round(np.arange(start, end, 0.1), 2))
plt.xlabel("FPR(1-TNR)")
plt.ylabel("TPR(Recall)")
plt.grid()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(
    y_test, 
    model.predict_proba(X_test), 
    multi_class='ovo',
    average="macro")

roc_auc
```
